```python
import tensorflow as tf
```

```python
def smartPrinter(arg1):
     print("---/\---")
     print(sess.run(arg1,feed_dict = feed_dict))
     print("---\/---")
    
with tf.Graph().as_default() as g:
    with tf.Session() as sess:
        input = tf.placeholder(dtype=tf.float32)
        output = tf.placeholder(dtype=tf.float32)
        raw_input_data = [[10,7],[5,4],[8,3],[2,8],[4,8],[9,1],[6,8],[1,7],[5,3],[7,6]]
        raw_output_data =[[97],[71],[74],[78],[90],[68],[95],[65],[70],[82]]
        feed_dict = {input:raw_input_data,output:raw_output_data}
        num_neurons_1 = 2
        num_neurons_2 = 3
        num_neurons_3 = 1
        #print(lst_input)
        normalized_input = input / tf.reduce_max(input,0)
        normalized_output = output / tf.reduce_max(output,0)
       # m_1 = tf.reduce_max(input,0)
       # m_2 = tf.reduce_max(output,0)
       # print(m_1)
       # print(m_2)
        weights_1 = tf.Variable(tf.random_normal([2,3]))
        bias_1 = tf.Variable(tf.random_normal([3]))
        
        weighted_sum = tf.matmul(normalized_input,weights_1) + bias_1
        
        activation_1 = tf.sigmoid(weighted_sum)
        
        weights_2 = tf.Variable(tf.random_normal([3,1]))
        bias_2 = tf.Variable(0.0)
        
        weighted_sum_2 = tf.matmul(activation_1,weights_2) + bias_2
        
        activation_2 = tf.sigmoid(weighted_sum_2)
        
        loss = tf.reduce_sum((activation_2 - normalized_output)**2)/10
        
        training_step = tf.train.GradientDescentOptimizer(learning_rate=0.02).minimize(loss)
        
        sess.run(tf.global_variables_initializer())
        smartPrinter(weights_1)
        smartPrinter(weights_2)
        for i in range(100):
            sess.run(training_step,feed_dict = feed_dict)

            smartPrinter(weights_1)
            smartPrinter(weights_2)
        smartPrinter(loss)
        smartPrinter(normalized_output)
        smartPrinter(activation_2)
        #print(sess.run(weighted_sum))
```
